## preview_data

Here we'll loop over all the GTFS files and load them into a [pandas](https://pandas.pydata.org/about.html) dataframe
so we can see what the data looks like.


```python
import os

import pandas
```

```python
data_directory = '../data/ST-GTFS'

files = []
for filename in os.listdir(data_directory):
    if filename.endswith(".txt") or filename.endswith(".csv"):
        filepath = os.path.join(data_directory, filename)
        print(filepath)
        files.append(filepath)
        continue
```
